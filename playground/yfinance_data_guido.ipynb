### Cambio porcentual por *t* y gráficos
pandas.DataFrame.pct_change<br>
pandas.DataFrame.resample

```python
ticker = "GGAL.BA"
data = yf.download(ticker, period='20y')
display('\n--Describe--\n', data.describe())
display('\n--Head--\n', data.head(4))
display('\n--Columns--\n', data.columns)

plt.style.use('dark_background')
plt.rcParams['figure.figsize'] = [12.0, 5]

# pandas.DataFrame.pct_change -> Computes the percentage change from the immediately previous row by default. This is useful in comparing the percentage of change in a time series of elements.
# variación diaria del precio de cierre
variaciones =  data['Adj Close'].pct_change()*100

display('\n--Variaciones--\n',variaciones)

agrupados = variaciones.groupby(data.index.year).sum()

display('\n--Variaciones agrupadas por año--\n',agrupados)

agrupados.plot(kind='bar',title='Rendimientos GGAL USD')
plt.show()
```

```python
# pandas.DataFrame.resample
# Convenience method for frequency conversion and resampling of time series. Object must have a datetime-like index

display(data['Adj Close'])
display(data['Adj Close'].resample('W').last())

variaciones = data['Adj Close'].pct_change()*100
agrupados = variaciones.groupby(data.index.week).mean()
agrupados.plot(kind='bar',title='Rendimientos GGAL USD')
plt.show()
```

```python
import pandas as pd
import pyfolio as pf

cartera = ['YPF','GGAL','AMZN','AAPL','TLT','SHY','IEF']
df = pd.DataFrame(columns=cartera)

for ticker in cartera:
    df[ticker] = yf.download(ticker, period='1y')['Adj Close']
```

```python
df
```

```python
# Percentage change between the current and a prior element.
df_new = df.pct_change().dropna().mean(axis=1)
display(df_new.describe())
```

```python
# Generate a number of tear sheets that are useful for analyzing a strategy's performance.
pf.create_full_tear_sheet(df_new)
```
